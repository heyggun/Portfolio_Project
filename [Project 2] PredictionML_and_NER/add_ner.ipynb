```python
data = pd.read_csv('member_conts.csv')
print(data.shape)
display(data.head(3))
```

```python
from konlpy.tag import Mecab
```

```python
mecab = Mecab()
```

```python
from tqdm import tqdm
```

```python
data['conts_nng'] = ''
stopwords = ['가입', '안녕', '남성', '여성', '나머지', '탈퇴', '여자', '남자', '카톡']
```

```python
for idx, row in tqdm(data.iterrows()):
    text = row['text']    
    try : 
        pos = mecab.pos(text)
        words = [n for n, tag in pos if tag in ['NNG'] if 2 <= len(n) <= 4 if n not in stopwords]
        words = list(set(words))
        
    except:
        sent = list()
        words = sent
    
    data.at[idx, 'conts_nng'] = words

data
```

```python
mate_nng_df = data[['conts_nng']]
mate_nng_df
```

```python
total_nngs = []
for row, idx in tqdm(mate_nng_df.iterrows()):
    nngs = idx['conts_nng']
    total_nngs = total_nngs + nngs
    total_nngs = list(set(total_nngs))
    
total_nngs
```

```python
def mecab_tokenize(sent):
    stopwords = pd.read_csv('stopwords.csv')
    mecab = Mecab()
    delwords = stopwords['stopwords'].tolist()

    try:
        pos = mecab.pos(sent)
        words = [n for n, tag in pos if tag in ['NNG'] if 2 <= len(n) <= 4 if n not in delwords]
        words = list(set(words))
    except:
        sent = list()
        words = sent
    return words
```
